# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
import math

def Radius(r):
    return (2/3) * math.pi * r**3

for r in range(2, 21, 2 ):
    volume = Radius(r)
    print(f"R: {r}, V: {volume:.2f}")
```

# Question 2 (20 points)

(a) Write a Python program to generate a sequence of 12 random numbers falling within the range of 50 to 150.

(b) Arrange the generated numbers in ascending order. Display the sorted list.

(c) Extend the program to arrange the numbers in descending order as well. Present the sorted list accordingly.

(d) Compute and output the average of the entire group of generated numbers.

(e) Enhance the program to calculate and display the average of the first 6 numbers from both the ascending and descending sorted lists.

```python
import random

# Making a list of twelve random numbers in between 50 to 150
num = [random.randint(50, 150) for _ in range(12)]
print(num)

print("\n")

# Arranging in ascending order
asc = sorted(num)
print("Numbers in ascending order:", asc)

print("\n")

# SArranging in descending order
desc = sorted(num, reverse=True)
print("Numbers in descending order:", desc)

print("\n")

# Calculating the avg of generated numbers
avg = sum(num) / len(num)
print("avg of the numbers:", avg)

print("\n")

# Calculating the avg of first 6 ascending order numbers
asc_avg = sum(asc[:6]) / 6
print("avg of the first 6 ascending order numbers:", asc_avg)

print("\n")

# Calculating the avg of first 6 descending order numbers
desc_avg = sum(desc[:6]) / 6
print("avg of the first 6 descending order numbers :", desc_avg)
```

# Question 3 (20 points).

Create a Python program that encompasses various string manipulation task:

(a) Prompt the user to input two sentences.

(b) Transform each sentence into a list, where each element corresponds to a word in the sentence. For example, if the sentences are "Hello, how are you doing today?" and "This is an example sentence for the program!"

The program should output:
['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(c) Compare the two sentences. Print out a list of words that occur in both sentences.

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python
import re
from collections import Counter
# Prompting to enter two sentences
A = input("first sentenc: ")
B = input("second sentence: ")

# Dividing sentence into series
list_A = re.findall(r'\w+|[^\s\w]', A)
list_B = re.findall(r'\w+|[^\s\w]', B)

# combining lists
c = list_A + list_B
print(c)

print("\n")

# Dividing sentences to words
list_A = re.findall(r'\w+', A)
list_B = re.findall(r'\w+', B)

# Finding common words
common= list(set(list_A) & set(list_B))

# Displaying combined list words
print("Words in both sentences:", common)

print("\n")

# erasing specified punctuations from both lists
spec_A = re.sub(r'[,.:;]', '', A)
spec_B = re.sub(r'[,.:;]', '', B)

# specified sentences
print("Specified first sentence:", spec_A)
print("Specified second sentence:", spec_B)

print("\n")

# vowelcount functiomn
def vow_cnt(text):
    return Counter(c for c in text.lower() if c in 'aeiou')

# Count vowels in each sentence
vow_A = vow_cnt(A)
vow_B = vow_cnt(B)

# Vowel count conversion and display function
def print_vowel_counts(vowel_counts, label):
    print(f"{label}: " + ", ".join(f"{v}:{vowel_counts.get(v, 0)}" for v in 'aeiou'))

# Print results
print_vowel_counts(vow_A, 'vowels in Sentence1')
print_vowel_counts(vow_B, 'vowels in Sentence2')
```

# Question 4 (20 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import math

def evaluate_expression(exp):
# for square root calculation swapping sqrt' with 'math.sqrt'
    exp = exp.replace('sqrt', 'math.sqrt')

    try:
        # Aseesing mathematical expression
        result = eval(exp, {"math": math})
        return result
    except ZeroDivisionError:
        return "Error: Cannot Divide by Zero"
    except SyntaxError:
        return "Error: Invalid syntax in expression."
    except NameError:
        return "Error: Invalid function"
    except Exception as e:
        # Catch any other exceptions
        return f"Error: {e}"

print("Hello")
print("Type 'exit' to quit when you are done.")

while True:
    # Prompting the user to enter a mathematical expression
    exp = input("Enter a mathematical expression: ")

    # Check for exit commands
    if exp.lower() in ['exit']:
        print("Thank You!!!!")
        break

    # Evaluate the expression and print the result
    result = evaluate_expression(exp)
    print("Result:", result)
```

# Question 5 (30 points).

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

To answer this question, you will need to download the file [Assignment1_denton_housing.csv](https://raw.githubusercontent.com/unt-iialab/info5731-spring2022/main/assignments/Assignment1_denton_housing.csv) included with this assignment. The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (10 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (10 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd

# Loading the Excel file
file_path = '/content/Assignment1_denton_housing.csv'
dt = pd.read_csv("/content/Assignment1_denton_housing.csv")

# Filtering data to get 'Occupied Housing Units' for every year
occu_units = dt[dt['title_field'] == 'Occupied Housing Units']

# Sorting data in ascending order
occu_units = occu_units.sort_values(by='year')

# differences calculation from year to year for number of occupied units
occu_units['difference'] = occu_units['value'].diff().fillna(0)
occu_units = occu_units.iloc[1:]

# Sorting the  differences in ascending order
occu_units = occu_units.sort_values(by='difference')

# Print resullts
for index, row in occu_units.iterrows():
    print(f" {row['year']-1} to {row['year']}: {row['difference']:+} units")

print("\n")

# Gathering the total, vacant, and occupied housing units for each year
tot_units = dt[dt['title_field'] == 'Total Housing Units']
vac_units = dt[dt['title_field'] == 'Vacant Housing Units']
occu_units = dt[dt['title_field'] == 'Occupied Housing Units']

# Merging year column data into single dataframe
merg_dt = pd.merge(tot_units, vac_units, on='year', suffixes=('_total', '_vacant'))
merg_dt = pd.merge(merg_dt, occu_units, on='year')

# percentage calculation
merg_dt['Vacant (%)'] = (merg_dt['value_vacant'] / merg_dt['value_total']) * 100
merg_dt['Occupied (%)'] = (merg_dt['value'] / merg_dt['value_total']) * 100

# Sorting data in ascending order
merg_dt = merg_dt.sort_values(by='year')

# Print results
for index, row in merg_dt.iterrows():
    print(f"{row['year']}   {row['Vacant (%)']:.2f}%   {row['Occupied (%)']:.2f}%")

print("\n")

# Filtering data to get 'Vacant Housing Units' and 'Occupied Housing Units'
vac_units = dt[dt['title_field'] == 'Vacant Housing Units']
occu_units = dt[dt['title_field'] == 'Occupied Housing Units']

# maximum vacant and occupied housing units
max_vacant = vac_units.loc[vac_units['value'].idxmax()]
max_occupied = occu_units.loc[occu_units['value'].idxmax()]

# Print results
print(f"{'Category':<100} {'Year':>2000} {'Value':>100}")
print(f"{'-'*75}")
print(f"{'Vacant Housing Units':<30} {int(max_vacant['year']):<10} {int(max_vacant['value']):<10}")
print(f"{'Occupied Housing Units':<30} {int(max_occupied['year']):<10} {int(max_occupied['value']):<10}")

print("\n")

def harmonic_mean(data):
    a = len(data)
    b = sum(1 / x for x in data)
    return a / b

# gathered inputs
data  = [5, 8, 12, 15, 20]

# Calculate harmonic mean
mean = harmonic_mean(data)
print(f"Harmonic mean is: {mean:.4f}")
```
